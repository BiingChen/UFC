## Import V1_Fight_Fighters_FighterInfo_w_flipped and convert columns to lowercase

```python
V1_df = pd.read_csv('../../02_Data/02_Processed_Data/V1_Fight_Fighters_FighterInfo_w_flipped.csv', index_col=0)
V1_df = V1_df.reset_index().drop(columns='index')
V1_df.columns = [col.lower() for col in V1_df.columns]
```

```python
# Build out the list of columns I want to pull from V1_df
stats_columns = ['f1_fighterid', 'f1_height','f1_weight','f1_stance']


# Create a dataframe fighter_stats
fighter_static_stats = V1_df[stats_columns].copy()
# Change Stance to boolean
fighter_static_stats.f1_stance = fighter_static_stats.f1_stance.map(lambda x: 0 if x == 'Orthodox' else 1)
# Select the mode for each value for each fighter since sometimes change weight class
fighter_static_stats = fighter_static_stats.groupby('f1_fighterid').agg(lambda x:x.value_counts().index[0]).reset_index()
# Currently some of the weights are zero.  Waiting for scrape to finish
fighter_static_stats.loc[fighter_static_stats.f1_weight ==0,'f1_weight'] = 185
# Since this data represents all fights, I will drop the "f1" from the column names
fighter_static_stats.columns = [col[3:] for col in fighter_static_stats.columns]
```
