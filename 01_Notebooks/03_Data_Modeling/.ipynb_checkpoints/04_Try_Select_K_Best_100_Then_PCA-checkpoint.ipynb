## Imports

```python
import pandas as pd
import pickle
from sklearn.decomposition import PCA
from matplotlib import pyplot as plt
import numpy as np
from sklearn.feature_selection import SelectKBest
import seaborn as sns
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegressionCV, LogisticRegression
import pickle
```

## Pull in training data

```python
with open('../../02_Data/02_Processed_Data/X_train.pkl', 'rb') as f:
    X_train = pickle.load(f)
    
with open('../../02_Data/02_Processed_Data/y_train.pkl', 'rb') as f:
    y_train = pickle.load(f)    

with open('../../02_Data/02_Processed_Data/X_test.pkl', 'rb') as f:
    X_test = pickle.load(f)
    
with open('../../02_Data/02_Processed_Data/y_test.pkl', 'rb') as f:
    y_test = pickle.load(f)    
```

## Try K Best

```python

selector = SelectKBest(k=100)
selector.fit(X_train, y_train)
col_mask = selector.get_support(indices=False)
# Create new dataframe with only desired columns, or overwrite existing
selected_columns = X_train.loc[:,col_mask].columns
```

## PCA for X_train on 100 selected columns

```python
components = 11
pca_r = PCA(n_components=components)
X_train_pca = pca_r.fit_transform(X_train[selected_columns])

plt.plot(np.cumsum(pca_r.explained_variance_ratio_))
plt.bar(range(components),pca_r.explained_variance_ratio_)
```

```python
X_train_pca
```

```python
components_loading_mat = pd.DataFrame(pca_r.components_, columns=X_train[selected_columns].columns, 
                                      index=[f'component_{x}' for x in range(pca_r.components_.shape[0])]).T
```

```python
components_loading_mat.sort_values('component_1_abs', ascending=False)
```

```python
for col in components_loading_mat.columns:
    components_loading_mat[f'{col}_abs'] = components_loading_mat[col].map(abs)
```

## Try to put the PCA into a model...?

```python
components = 11
pca = PCA(n_components=components)
X_train_pca = pca.fit_transform(X_train[selected_columns])
X_test_pca = pca.transform(X_test[selected_columns])
```

```python
lr = LogisticRegression(C=10)
lr.fit(X_train_pca, y_train)
lr.score(X_train_pca, y_train),lr.score(X_test_pca, y_test)
```
